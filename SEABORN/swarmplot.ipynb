```python
sns.boxplot(data=cars,x="origin",y="horsepower",
           whis=np.inf)
sns.swarmplot(data=cars,x="origin",y="horsepower",
             color="fuchsia",alpha=0.40)
```

```python
sns.violinplot(data=cars,x="origin",y="horsepower",
           whis=np.inf)
sns.swarmplot(data=cars,x="origin",y="horsepower",
             color="cyan",alpha=0.5)
```

```python
plt.figure(figsize=(12,7))
ax = sns.violinplot(x="origin",y="horsepower",data=cars,
                    color='fuchsia')


num_items = len(ax.collections)
sns.stripplot(x="origin",y="horsepower",data=cars,
              palette=['blue', 'deepskyblue',"r"], alpha=0.4, size=7)

for item in ax.collections[num_items:]:
    item.set_offsets(item.get_offsets() + 0.15)
    
    
sns.boxplot(x="origin",y="horsepower",data=cars,
            width=0.25,
            showfliers=False,
            showmeans=True, 
            meanprops=dict(marker='o', markerfacecolor='darkorange',
                           markersize=10, zorder=3),
            boxprops=dict(facecolor=(1,.9,.2,0), 
                          linewidth=3, zorder=3),
            whiskerprops=dict(linewidth=3),
            capprops=dict(linewidth=3),
            medianprops=dict(linewidth=3,color="lime"))
```
