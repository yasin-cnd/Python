```python
geyser = sns.load_dataset("geyser")
```

```python
geyser.head()
```

```python
sns.kdeplot(data=geyser,x="duration",y="waiting")
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",hue="kind")
```

```python
#fill
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    hue="kind",
    fill=True)
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    hue="kind",
    fill=False)
```

```python
#levels
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    hue="kind",
    fill=False,
    levels=5)
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    hue="kind",
    fill=False,
    levels=10)
```

```python
#cmap
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    fill=True,
    thresh=0,
    levels=10,
    cmap="viridis")
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    fill=True,
    thresh=0,
    levels=10,
    cmap="tab20b")
```

```python
#cbar
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    fill=True,
    thresh=0,
    levels=10,
    cmap="viridis",
    cbar=True)
```

```python
sns.kdeplot(data=geyser,y="waiting",x="duration",
    fill=True,
    thresh=0,
    levels=10,
    cmap="viridis",
    cbar=False)
```
