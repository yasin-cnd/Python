```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
diamonds.info()
```

```python
diamonds=diamonds.sample(n=200,random_state=44)
```

```python
#set_style  white, dark, whitegrid, darkgrid, ticks
```

```python
sns.set_style("ticks")
```

```python
sns.regplot(data=diamonds,x="carat",y="price")
```

```python
sns.set_style("dark")
```

```python
#fit_reg
```

```python
 sns.regplot(data=diamonds,x="carat",y="price",
            fit_reg=True) 
```

```python
 sns.regplot(data=diamonds,x="carat",y="price",
            fit_reg=False) 
```
