```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
cars=sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.describe()
```

```python
xlabels = ['ocak', 'şubat', 'mart', 'nisan', 'mayıs']
ylabels = ['kahve', 'çay', 'su', 'soda']

orders = np.array([[25,20,30,22,30], 
                   [60,45,21,18,15],
                   [25,24,30,12,12],
                   [32,20,11,20,11]])
```

```python
#cmap
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(orders, 
            cmap='Accent',
            vmin=0,
            xticklabels=xlabels,
            yticklabels=ylabels,
            annot=True,
            square=True,
            annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'}
           )
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(orders, 
            cmap='PuBu',
            vmin=0,
            xticklabels=xlabels,
            yticklabels=ylabels,
            annot=True,
            square=True,
            annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'})
```
