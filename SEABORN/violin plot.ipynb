```python
import seaborn as sns
```

```python
cars=sns.load_dataset("mpg").dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(data=cars,x="displacement")
```

```python
sns.violinplot(data=cars,y="displacement",x="cylinders",
              hue="origin")
```

```python
sns.violinplot(data=cars[cars.origin.isin(["japan","europe"])],
              x="cylinders",y="displacement",
              hue="origin")
```
