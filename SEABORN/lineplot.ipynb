```python
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event",
             style="event",
            markers=True)
```

```python
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event",
             style="event",
            markers=False)
```

```python
#dashes
```

```python
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event",
             style="event",dashes=True)
```

```python
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event",
             style="event",dashes=False)
```

```python
#err_style
```

```python
sns.lineplot(data=fmri,x="timepoint",y="signal",hue="event",
    err_style="bars")
```

```python
sns.lineplot(data=fmri,x="timepoint",y="signal",hue="event",
    err_style="band")
```

```python
# ci
```

```python
sns.lineplot(
    data=fmri, x="timepoint", y="signal", hue="event",
    ci=77)
```

```python
sns.lineplot(
    data=fmri, x="timepoint", y="signal", hue="event",
    ci=100)
```

```python
#lw
```

```python
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
```

```python
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
     estimator=None,lw=2.5,
)
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
```

```python
dots.head()
```

```python
#legend
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate", hue="coherence",
             style="choice",
             legend=False)
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate", hue="coherence",
             style="choice",
             legend=True)
```

```python
# palette = deep , tab10 , hls , husl , Set2 , rocket , mako ,
#crest , viridis , dark:salmon_r , coolwarm
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate",
    hue="coherence", style="choice",
    palette="husl")
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate",
    hue="coherence", style="choice",
    palette="Set2")
```

```python
#sizes
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate",
    size="coherence", hue="choice",
    sizes=(2,5))
```

```python
sns.lineplot(data=dots, x="time", y="firing_rate",
    size="coherence", hue="choice",
    sizes=(.2,4))
```
